Assigning the numeric value in seniority 

```python
df['seniority']=df['seniority'].map({'jr':0, 'na':-1, 'senior':1})
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

bar plot ---seniority vs Avg salary


```python
sns.barplot(x='seniority',y='avg_salary',data=df)
plt.title("salary trend by experience")
plt.show()
```

```python
skills=['python',
       'excel', 'hadoop', 'spark', 'aws', 'tableau', 'big_data']
skill_counts=df[skills].sum().sort_values(ascending=False)
skill_counts.values
```

skill demand trend means how many jobs want which skills

```python
sns.barplot(x=skill_counts.index,y=skill_counts.values)
plt.ylabel("Number of Jobs Requiring Skill")
plt.title("skill demand trend")
```
