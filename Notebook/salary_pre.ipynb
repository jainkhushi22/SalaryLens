```python
data=df.drop([ 'Salary Estimate', 'Job Description', 
       'Company Name','Location', 'Headquarters', 'Size', 'Type of ownership',
       'Industry','Revenue', 'min_salary', 'max_salary',
      'job_state', 'same_state', 'company_age',  'job_simp'],axis=1)
```

```python
data['seniority']=data['seniority'].map({'senior':2, 'na':0, 'jr':1})
```

```python
df['avg_salary'].unique().shape
```

Predicting salaries based on job postings scraped from Glassdoor

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
job=le.fit_transform(data['Job Title'])
sec=le.fit_transform(data['Sector'])
```

```python
data['Job Title']=job
data['Sector']=sec
```

```python
data.head(3)
```

Training the model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data.drop(['avg_salary'],axis=1),data['avg_salary'],test_size=0.3,random_state=30)
```

Lasso and Ridge because linear regression is overfitting (we have large data)

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
model=Lasso(alpha=1.5)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model2=Ridge(alpha=0.7)
model2.fit(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model2=DecisionTreeClassifier(max_depth=8)
model2.fit(X_train,y_train)
```

```python
model2.score(X_train,y_train)
```

```python
model2.score(X_test,y_test)
```

Randomforest classifier this is best because of handing Outliers

```python
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier(n_estimators=700,max_depth=17,min_samples_leaf=4)
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
model1.score(X_train,y_train)
```
